# Importing the Libraries

```python
import ollama
from ollama import Client
```

# Setting up the Local Clent

```python
client = Client(host='http://localhost:11434')
```

# Creating the Main Responce Function with Simple Memory

### Each of the previous question and answers are appended to the next input.
### GPT-oss which is an open source Chat GPT model released by OpenAI is used in this example.

```python
messages = []

def model(question):
    user_input = {'role': 'user', 'content': question}
    messages.append(user_input)

    response = ollama.chat(model='gpt-oss:20b', messages=messages)
    print('ANS:', response['message']['content'] )
    messages.append({'role': 'assistant', 'content': response['message']['content']})
```

# Executng the function to replicate a chat bot

```python
print('Enter your Question or enter EXIT: \n')
while True :
    question  = input('USER:' )
    if question == 'EXIT':
        break
    model(question)
    print('\n')
```
